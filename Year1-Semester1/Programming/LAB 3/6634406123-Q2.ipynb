## Q2 (7 points)
จงเขียนโปรแกรมที่รับน้ำหนัก (kg.) และส่วนสูง (m.) จากผู้ใช้มาคำนวณ bmi จาก น้ำหนัก (kg.)/ส่วนสูง
(m.) ยกกำลัง 2 แล้วแสดงผลลัพธ์ว่าผู้ใช้มีรูปร่างในระดับใดดังนี้

ต่ำกว่า 18.5 ผอม

ไม่น้อยกว่า 18.5 แต่ต่ำกว่า 23.0 รูปร่างปกติ

ไม่น้อยกว่า 23.0 แต่ต่ำกว่า 25.0 รูปร่างอ้วน

ไม่น้อยกว่า 25.0 แต่ต่ำกว่า 30.0 อ้วนระดับ 1

ไม่น้อยกว่า 30.0 อ้วนระดับ 2

### ตัวอย่างการทำงาน: 

Weight (kg.) : **61.3** <br>
Height (m.) : **1.65** <br>
รูปร่างปกติ

### ตัวอย่างการทำงาน: 

Weight (kg.) : **50.3** <br>
Height (m.) : **1.67**<br>
ผอม

### ตัวอย่างการทำงาน:

Weight (kg.) : **72.3** <br> Height (m.) : 1.5<br>อ้วนระดับ 2


```python
def run():
    # YOUR CODE HERE
    Weight = float(input("Weight (kg.) : "))
    Height = float(input("Height (m.) : "))
    sum = float(Weight / Height**2)
    if sum<18.5 :
        print ("ผอม")
    elif sum>=18.5 and sum<23.0 :
        print ("รูปร่างปกติ")
    elif sum>=23.0 and sum<25.0 :
        print ("รูปร่างอ้วน")
    elif sum>=25.0 and sum<30.0 :
        print ("อ้วนระดับ 1")
    elif sum>=30.0 :
        print ("อ้วนระดับ 2")
```

```python
'''ตรวจสอบผลลัพธ์ของโปรแกรมด้วยการรันที่ cell นี้'''
run()
```

```python
'''ตรวจสอบผลการคำนวณ'''
from unittest.mock import patch
from io import StringIO
import math

def test(inputs,expected_output): 
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert mock_stdout.getvalue() == expected_output,mock_stdout.getvalue()
def test_easy_print(inputs,expected_output):
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert expected_output in mock_stdout.getvalue(), mock_stdout.getvalue()
```

```python
'''ตรวจสอบการแสดงผลลัพธ์อย่างง่าย'''
test_easy_print(inputs=['61.3','1.65'],expected_output='รูปร่างปกติ')
```
