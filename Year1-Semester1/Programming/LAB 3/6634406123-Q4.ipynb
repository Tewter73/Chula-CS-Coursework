จงเขียนโปรแกรมที่รับรหัสนิสิต แล้วตรวจสอบว่าเป็นรหัสที่ใช้ได้ คือ

- เป็นเลข 10 หลัก

- เลข 2 หลักแรกเป็นปีที่เข้าเรียน (50-66)

- เลขหลักที่ 3 (จากหน้า) เป็น 3, 4 หรือ 7

- เลข 2 หลักสุดท้ายของรหัสนิสิตเป็นรหัสคณะ คือ 21 22 23 24 25 26 27 28

### ตัวอย่างการทำงาน: 

Enter student ID : **6033235423** <br> Valid ID

### ตัวอย่างการทำงาน: 

Enter student ID : **6070013223** <br> Valid ID

### ตัวอย่างการทำงาน:

###  ตัวอย่างการทำงาน:

### ตัวอย่างการทำงาน:

### ตัวอย่างการทำงาน: 

Enter student ID : **2013235499** <br> Invalid ID

```python
def run():
    # YOUR CODE HERE
    studentID = (input("Enter student ID : "))
    if not studentID.isdigit() or len(studentID) != 10 :
        print ("Invalid ID")
    elif not 50 <= int(studentID[:2]) <= 66 :
        print ("Invalid ID")
    elif studentID[2] not in ("3","4","7") :
        rint ("Invalid ID")
    elif not 21 <= int(studentID[-2:]) <= 28 :
        print ("Invalid ID")
    else :
        print ("Valid ID")
```

```python
'''ตรวจสอบผลลัพธ์ของโปรแกรมด้วยการรันที่ cell นี้'''
run()
```

```python
'''ตรวจสอบผลการคำนวณ'''
from unittest.mock import patch
from io import StringIO
import math

def test(inputs,expected_output): 
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert mock_stdout.getvalue() == expected_output,mock_stdout.getvalue()
def test_easy_print(inputs,expected_output):
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert expected_output in mock_stdout.getvalue(), mock_stdout.getvalue()
```

```python
'''ตรวจสอบการแสดงผลลัพธ์อย่างง่าย'''
test_easy_print(inputs=['6633235423'],expected_output='Valid')
```
