## Methos of file objects

Method - Description
*   close()- Closes the file
*   detach()- Returns the separated raw stream from the buffer
*   fileno() - Returns a number that represents the stream, from the operating system's perspective
*   flush()	- Flushes the internal buffer
*   isatty()- Returns whether the file stream is interactive or not
*   read()	- Returns the file content
*   readable()	- Returns whether the file stream can be read or not
*   readline()	- Returns one line from the file
*   readlines()	Returns a list of lines from the file
*   seek()	- Change the file position
*   seekable()	- Returns whether the file allows us to change the file position
*   tell()	- Returns the current file position
*   truncate()	- Resizes the file to a specified size
*   writable()	- Returns whether the file can be written to or not
*   write()	- Writes the specified string to the file
*   writelines()	- Writes a list of strings to the file

## Creating your own functions

### Writing a basic function

The general structure to create your own function.

def functionName(parameter list):  
> statement list





````python
def <function_name>(<param1>,<param2>,...):
  <statement>
  <statement>
  ...
return <returning_value>  #this line is optional for void functions, but required for returning-value functions
````

* The first line,  begins with `def` to define a function block, followed by the function name, which must comply with the naming rules as used for variables, followed by a set of parameter(s) within the parentheses, and ends with a colon `:`.
* A function can have zero, one or more parameters or arguments
>
> no arguments - write empty parentheses `()`
>
> more than one arguments - separated by a comma `,`
* The next lines contain a set of statements that will be executed anytime when the function is called.
These statement are indented and belong to the function as a group known as a <font color=red>function block</font>



### How a program with functions works

* Before calling a function, we need to first run the function block so that the interpreter knows this function exists and what it is defined.
* It is important to note that running the entire function block does not execute the function.
* The fucntion will perform the task inside when it is called.
* Like built-in functions, can call a function anywhere in program by its name, followed by the parentheses in which all the required parameters are passed.

### Local vs Global variables

* The variables created inside a function, called <font color=red>local variables</font>, belong to the function only.
* These variables cannot be accessed or used by any statment outside the function. --> the scope of these variables is within the function block.
* When the function ends, local variables will return the memory space reserved.
* In case of having the variable with the same name created outside the function, remember that they are two different variables.

````python
def sum_to(n):
    sum = 0
    i = 1
    while i <= n:
        sum = sum+i
        i = i+1
    print('Value of sum inside the function:',sum)
    return sum

# Main program ก็คือถ้าไม่ใช่ global ถ้าเรากำหนดตัวแปรใหม่นอกฟัง์ชัน มันก็จะใช้ตัวแปรใหม่
sum = 5
sum_to(5)
print('Value of sum outside the function:',sum)
````

* The values of `sum` inside and outside the function are different.
* Calling the function `sum_to` that changes the value of sum internally does not affect the value of sum outside the function at all.


* However, if we want to use the variable defined outside the funciton *(but in the main program)*, called <font color=red>global variables</font> within the function, we can have the statement, `global <variable>` before using such variable inside the function.


````python
def sum_to(n):
    global sum # This line is to say that the variable is inside and outside the function are the same variable.
    print('Inside this function before doing anything, sum has a value of', sum)
    sum = 0
    i = 1
    while i <= n:
        sum = sum+i
        i = i+1
    return sum

# Main program
sum = 5
sum_to(5)
print('After calling the function, sum has a value of', sum)
````

* With the `global sum` statement within the function, the values of `sum` outside the function was changed because the function `sum_to` had change the value of sum.

### Using global mutable variables in functions

````python
a_list = [1,2,3,4]

def change(pos,value):
    a_list[pos] = value

change(0,-1)
print(a_list)
````

This parameter *'pos'* is connected to the global mutable variables, so if you modify its value within the function, it will also modify the value of that global mutable variables.

````python
a_list = [1,2,3,4]

def change(pos,value):
    a_list[pos] = value
    return a_list

a_list = change(0,-1)
print("first calling function a = ",a_list)

b_list = change(0,-2)
print("second calling function a = ",a_list)
print("b = ",b_list)
````
