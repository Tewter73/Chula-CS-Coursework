(ถือว่านักเรียนต้องตอบข้อสอบทุกข้อ)
และมีนักเรียนอยู่ 8 คน จากนั้นให้ตรวจคำตอบของนักเรียนทุกคน แล้วหา
- คะแนนของนักเรียนแต่ละคน
- คะแนนเฉลี่ยของข้อสอบแต่ละข้อ
- ข้อสอบข้อที่นักเรียนทำคะแนนได้น้อยสุด (แสดงให้ครบทุกข้อ หากคะแนนเท่ากันหลายข้อ )
- ข้อสอบข้อที่นักเรียนทำคะแนนได้มากสุด (แสดงให้ครบทุกข้อ หากคะแนนเท่ากันหลายข้อ )

**หมายเหตุ ให้ใช้ลิสต์ในข้อนี้ และสามารถใช้ฟังก์ชัน min และ max เพื่อหาค่าน้อยสุดและมากสุดในลิสต์ได้**

### ตัวอย่างการทำงาน:

Hardest : 3 <br>
Easiest : 2 6 <br>

```python
def run():
    # YOUR CODE HERE
    Solution_file = input('Choose your solution file: ')
    Exam_file = input('Choose your exam file: ')
    scores = []
    with open(Solution_file, 'r') as f:
        solution = f.readline().strip().split()

    with open(Exam_file, 'r') as f:
        E = []
        lines = f.readlines()
        for line in lines:
            E.append(line.strip().split())

    for i in range(len(E)):
        score = 0
        for j in range(len(solution)):
            if E[i][j] == solution[j]:
                score += 1
        scores.append(score)

    print(f'Student score : {scores}\n')

    average_scores = []
    for j in range(len(solution)):
        avg = 0.0
        for i in range(len(E)):
            if E[i][j] == solution[j]:
                avg += 0.125
        average_scores.append(avg)
        print(f'Question {j+1} : {average_scores[j]}')

    min_s = ''
    max_s = ''
    for i in range(len(average_scores)):
        if average_scores[i] == min(average_scores):
            s_min = str(i+1)
            min_s = min_s + ' ' + s_min
        if average_scores[i] == max(average_scores):
            s_max = str(i+1)
            max_s = max_s + ' ' + s_max
            
    print(f'\nHardest :{min_s}')
    print(f'Easiest :{max_s}')
```

```python
'''ตรวจสอบผลลัพธ์ของโปรแกรมด้วยการรันที่ cell นี้'''
run()
```

```python
'''ตรวจสอบผลการคำนวณ'''
from unittest.mock import patch
from io import StringIO

def test(inputs,expected_output): 
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert mock_stdout.getvalue() == expected_output,mock_stdout.getvalue()
def test_easy_print(inputs,expected_output):
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert expected_output in mock_stdout.getvalue(), mock_stdout.getvalue()
```
