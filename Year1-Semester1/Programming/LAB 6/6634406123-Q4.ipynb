จงเขียนโปรแกรมที่วนรับจำนวนเงินที่ต้องการเบิกจากบัญชี โดยจะยอมให้เบิกถ้ายังมีเงินในบัญชีพอ และวนรับจำนวนเงินที่ต้องการเบิกไปจนเงินในบัญชีหมดพอดี (กำหนดให้เริ่มต้นมีเงินในบัญชี 50,000 บาท)


### ตัวอย่างการทำงาน: 

withdraw : **50000** <br>
Balance is 0

### ตัวอย่างการทำงาน: 

withdraw : **45000**  <br>
withdraw : **5000**  <br>
Balance is 0

### ตัวอย่างการทำงาน: 

withdraw : **60000**  <br>
Insufficient fund. <br>
withdraw : **40000**  <br>
withdraw : **20000**  <br>
Insufficient fund.<br>
withdraw : **10000**  <br>
Balance is 0

```python
def run():
    # YOUR CODE HERE
    Balance = 50000
    Withdraw = int(input("withdraw : "))
    while Withdraw < Balance or Withdraw > Balance :
        if Withdraw <= Balance :
            Balance -= Withdraw
            Withdraw = int(input("withdraw : "))
        else :
            print("Insufficient fund.")
            Withdraw = int(input("withdraw : "))
    print("Balance is 0")
```

```python
'''ตรวจสอบผลลัพธ์ของโปรแกรมด้วยการรันที่ cell นี้'''
run()
```

```python
'''ตรวจสอบผลการคำนวณ'''
from unittest.mock import patch
from io import StringIO

def test(inputs,expected_output): 
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert mock_stdout.getvalue() == expected_output,mock_stdout.getvalue()
def test_easy_print(inputs,expected_output):
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert expected_output in mock_stdout.getvalue(), mock_stdout.getvalue()
```

```python
'''ตรวจสอบการแสดงผลลัพธ์อย่างง่าย'''
test_easy_print(inputs=['50000'],expected_output='Balance')
```
