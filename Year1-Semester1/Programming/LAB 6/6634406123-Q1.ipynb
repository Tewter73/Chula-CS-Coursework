จงเขียนโปรแกรมที่รับคำมาต่อกันเป็นประโยค โดยจะหยุดรับเมื่อรับ . มา

### ตัวอย่างการทำงาน: 

Next word : **.** <br>
Sentence:

### ตัวอย่างการทำงาน: 

Next word : **Bye** <br>
Next word : **bye** <br>
Next word : **.** <br>
Sentence: Bye bye

### ตัวอย่างการทำงาน: 

Next word : **My** <br>
Next word : **name** <br>
Next word : **is** <br>
Next word : **next** <br>
Next word : **word** <br>
Next word : **.** <br>
Sentence: My name is next word

### ตัวอย่างการทำงาน: 

Next word : **My** <br>
Next word : **name** <br>
Next word : **is** <br>
Next word : **Chicky** <br>
Next word : **Chicky!** <br>
Next word : **.** <br>
Sentence: My name is Chicky Chicky!

```python
def run():
    # YOUR CODE HERE
    Nword = input("Next word : ")
    Sentence = ""
    while Nword != "." :
        Sentence = Sentence + " " + Nword
        Nword = input("Next word : ")
    print(f"Sentence:{Sentence}")
```

```python
'''ตรวจสอบผลลัพธ์ของโปรแกรมด้วยการรันที่ cell นี้'''
run()
```

```python
'''ตรวจสอบผลการคำนวณ'''
from unittest.mock import patch
from io import StringIO

def test(inputs,expected_output): 
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert mock_stdout.getvalue() == expected_output,mock_stdout.getvalue()
def test_easy_print(inputs,expected_output):
    with patch('builtins.input', side_effect=inputs), patch('sys.stdout', new_callable=StringIO) as mock_stdout:
        run()
        assert expected_output in mock_stdout.getvalue(), mock_stdout.getvalue()
```

```python
'''ตรวจสอบการแสดงผลลัพธ์อย่างง่าย'''
test_easy_print(inputs=['.'],expected_output='Sentence')
```

```python
'''ตรวจสอบผลการคำนวณและการแสดงผลลัพธ์ให้เหมือนกับโจทย์ที่กำหนดให้'''
test(['.'],'Sentence:\n') 
test(['Welcome','.'],'Sentence: Welcome\n')
test(['Comprog','.'],'Sentence: Comprog\n')
test(['Bye','bye','.'],'Sentence: Bye bye\n') 
test(['I','love','Comprog','.'],'Sentence: I love Comprog\n')
test(['My','name','is','next','word','.'],'Sentence: My name is next word\n')  
test(['My','name','is','Chicky','Chicky!','.'],'Sentence: My name is Chicky Chicky!\n')
```
